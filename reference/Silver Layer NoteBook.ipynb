#  # Transfermations

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

##### Date

```python

df_cal = df_cal.withColumn('Month', month (col('Date')))\
    .withColumn('Year', year (col('Date')))
```

##### Concat_ws & _Select_

```python
df_customers = df_customers.withColumn("FullName", concat_ws(" ", col("Prefix"), col("FirstName"), col("LastName")))

df_customers = df_customers.select(['CustomerKey', 'FullName', 'BirthDate', 'MaritalStatus', 'Gender', 'EmailAddress', 'AnnualIncome', 'TotalChildren', 'EducationLevel', 'Occupation', 'HomeOwner'])
```

##### Spilt

```python
df_products = df_products.withColumn('ProductSKU', split(col('ProductSKU'), ',') [0])\
                        .withColumn('ProductName', split(col('ProductName'), ' ') [0])
```

##### Time Stamp

```python
df_sales = df_sales.withColumn('StockDate', to_timestamp(col('StockDate')))
```

##### Regexp_Replace

```python
df_sales = df_sales.withColumn('OrderNumber', regexp_replace(col('OrderNumber'), r'\d', 'X'))
```

##### Aggregate

```python
df_sales = df_sales.withColumn('Multiply', col('OrderLineItem')*col('OrderQuantity'))
```
